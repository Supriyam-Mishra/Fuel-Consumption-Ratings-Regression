## Outliers

```python
df.info()
```

```python
c=[]
for i in df.columns:
    if df[i].dtypes!='object':
        c.append(i)
print(c)
```

```python
df.shape
```

```python
plt.figure(figsize=(20,20))
df[['Engine Size(L)', 'Cylinders', 'Fuel Consumption(Comb (L/100 km))', 'CO2 Emissions(g/km)', 'CO2 Rating', 'Smog Rating']].boxplot(vert=False)
plt.show()
```

```python
for i in c:
    q25=df[i].quantile(0.25)
    q75=df[i].quantile(0.75)
    iqr=q75-q25
    print(f"{i} IQR:{iqr}")
    upper_limit=q75+(1.5*iqr)
    print(f"{i} upper_limit:{upper_limit}")
    lower_limit=q25-(1.5*iqr)
    print(f"{i} lower_limit:{lower_limit}")
    df=df[(df[i]>=lower_limit) & (df[i]<=upper_limit)]
```

## LabelEncoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
col=['Transmission','Fuel Type']
```

## Independent And Dependent features

```python
X=df.drop('Fuel Consumption(Comb (L/100 km))',axis=1)
y=df['Fuel Consumption(Comb (L/100 km))']
```

```python
X
```

```python
#### Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

## RandomForest

```python
from sklearn.ensemble import RandomForestRegressor
RFR = RandomForestRegressor()
RFR.fit(X_train, y_train)
prediction = RFR.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.mean_squared_error(y_test,prediction)
```

```python
metrics.r2_score(y_test,prediction)
```

```python
def adj_r2(r2, n, p): 
     return 1-((1-r2)*(n-1)/(n-p-1))



for i in range(45,120,1):
    for j in range(2,16,1):
        for k in range(10,30,1):
            rf = RandomForestRegressor(n_estimators = k, random_state=i,max_depth=j)

            rf.fit(X_train, y_train)

            trainadrj32 = adj_r2(rf.score(X_train, y_train), len(X_train), len(X_train.columns))
            testadrj32 = adj_r2(rf.score(X_test, y_test), len(X_test), len(X_test.columns))
            if (abs(trainadrj32 - testadrj32) < .01) and (trainadrj32 > .80):
                print(k, i, j)
                print('************** adj R2 Train: {} **************'.format(adj_r2(rf.score(X_train, y_train), len(X_train), len(X_train.columns))))
                print('************** adj R2 Test: {} **************'.format(adj_r2(rf.score(X_test, y_test), len(X_test), len(X_test.columns))))
                print('**************')
```
